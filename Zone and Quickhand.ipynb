```python
import numpy as np
import pandas as pd
import seaborn as sns

in_file = 'iOS_日_留存用户_20160401_20160531.csv'
ios_data = pd.read_csv(in_file)
new_ios_data=ios_data.drop(['2天后留存率','4天后留存率','5天后留存率','6天后留存率','14天后留存率'],axis=1)
new_ios_data.columns=['Date','New Users','1 survival rate','3 survival rate','7 survival rate','30 survival rate']
new_ios_data['Date']=new_ios_data['Date'].apply(lambda x:(x[5:7]+x[8:]))
new_ios_data.head()
```

```python
in_file = 'Android_日_留存用户_20160401_20160531.csv'
new_android_data=pd.read_csv(in_file)
new_android_data=new_android_data.drop(['2天后留存率','4天后留存率','5天后留存率','6天后留存率','14天后留存率'],axis=1)
new_android_data.columns=['Date','New Users','1 survival rate','3 survival rate','7 survival rate','30 survival rate']
new_android_data['Date']=new_android_data['Date'].apply(lambda x:(x[5:7]+x[8:]))
new_android_data.head()
```

```python
import matplotlib.pyplot as plt
%pylab inline
plt.figure(figsize=(12,12))
sns.pointplot(x='Date',y='1 survival rate',data=new_ios_data)
```

```python
import matplotlib.pyplot as plt
%pylab inline
plt.figure(figsize=(12,12))
sns.pointplot(x='Date',y='1 survival rate',data=new_android_data)
```
