```python
import numpy as np
import pandas as pd
import seaborn as sns

in_file = '(0925-1024)(All).csv'
full_data = pd.read_csv(in_file)
full_data.head()
```

```python
full_data.shape[0]
```

```python
data=full_data.drop(['新增独立','活跃独立'],axis=1)
data.head()
```

```python
data.columns=['Date','New Users','DAU','Numbers Of Starts','Cumulative Users']
data.head()
```

```python
data['Date']=data['Date'].apply(lambda x:(x[5:7]+x[8:]))
data.head()
```

```python
import matplotlib.pyplot as plt
%pylab inline
plt.figure(figsize=(12,12))
sns.pointplot(x="Date", y="DAU", data=data)
```

```python
import matplotlib.pyplot as plt
%pylab inline
plt.figure(figsize=(12,12))
sns.pointplot(x="Date", y="New Users", data=data)
```

```python
import matplotlib.pyplot as plt
%pylab inline
plt.figure(figsize=(12,12))
sns.pointplot(x="Date", y="Numbers Of Starts", data=data)
```
